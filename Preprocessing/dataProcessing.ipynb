````python
SPECIAL_IDS: list[str] = ["U", "UU", "UUU", "NN", "NNN", "30L", "25*", "CR", "N/A", "Y"]
````

````python
# Tách cột ID
def set_id(
    df: pd.DataFrame,
    col: str
) -> pd.DataFrame:
    
    new_df: pd.DataFrame = df.copy()
    
    def handle_value(value: Union[str, float]):
        if type(value) != str:
            return "_"
                     
        values: list[str] = value.split(" ", 1)
        _id: str = values[0]
        if not _id.isdigit() and _id not in SPECIAL_IDS:
            return "_"
        
        return _id
    
    new_df[f"{col} ID"] = new_df[col].apply(handle_value)
    
    return new_df

# Tách cột Desc
def set_desc(
    df: pd.DataFrame,
    col: str
) -> pd.DataFrame:

    new_df: pd.DataFrame = df.copy()

    def handle_value(value: Union[str, float]):
        if type(value) != str:
            return "_"
        
        values = value.split(" ", 1)
        if len(values) == 1:
            return "_"
                     
        _id: str = values[0]
        _desc: str = values[1].strip(" -.")
        if not _id.isdigit() and _id not in SPECIAL_IDS:
            half_length: int = len(value) // 2
            _desc = value[:half_length].strip(" -. ")
        
        return _desc
     
    new_df[f"{col} Desc"] = new_df[col].apply(handle_value)
    
    return new_df
````

````python
def get_id_with_value(
    grouped_df: pd.DataFrame,
    col: str,
    condition: pd.DataFrame.mask
) -> pd.DataFrame:
    return grouped_df.loc[
        condition, [f"{col} ID", f"{col} Desc"]
    ].drop_duplicates()
````

````python
# Kết bảng bằng ID
def merge_dfs(
    df_1: pd.DataFrame,
    df_2: pd.DataFrame,
    col: str
):
    merged_df = df_1.merge(
        df_2,
        on=[f"{col} ID"],
        suffixes=(" Old", " New")
    ).drop_duplicates()
    
    merged_df = merged_df[
        merged_df[f"{col} Desc Old"] !=
        merged_df[f"{col} Desc New"]
    ]
    
    return merged_df
````

````python
# Thay thế giá trị cũ bằng giá trị mới trong bảng
def replace_value_from_hash_df(
    df: pd.DataFrame,
    hash_df: pd.DataFrame,
    col: str,
    label: str,
    is_debug=True
):
    new_df = df.copy()
    
    for index, row in hash_df.iterrows():
        row_id = row[f"{label} ID"]
        row_old_desc = row[f"{label} Desc Old"]
        row_new_desc = row[f"{label} Desc New"]
        
        new_df.loc[
            (new_df[f"{col} ID"] == row_id) &
            (new_df[f"{col} ID"] != "_") &
            (new_df[f"{col} Desc"] != row_new_desc) &
            (new_df[f"{col} Desc"] != "_"),
            f"{col} Desc"
        ] = row_new_desc
        
        if is_debug:
            print(f"Handled '{col}'-'{row_id}': '{row_old_desc}' to '{row_new_desc}'")

        
    return new_df
````

## Xử lý ```Action Taken Primary```, ```Ignition Factor Primary```, ```Property Use```

````python
# Xử lý trên từng cột
def handle_each(
    df: pd.DataFrame,
    col: str
):
    new_df = df.copy()

    # Handle some dump ids in specific columns
    if col in ["Action Taken Primary", "Ignition Factor Primary"]:
        new_df[col + " ID"] = new_df[col + " ID"].replace('0', '00')
    elif col == "Property Use":
        new_df[col + " ID"] = new_df[col + " ID"].replace('0', '000')
        new_df[col + " ID"] = new_df[col + " ID"].replace('0000', '000')

    grouped_df = new_df.groupby([
        f"{col} ID",
        f"{col} Desc",
        "Year",
        "Month"
    ]).size().reset_index()

    id_with_old_value_df = get_id_with_value(
        grouped_df,
        col,
        ((grouped_df["Year"] <= 2013))
    )
    
    # print(id_with_old_value_df)

    id_with_new_value_df = get_id_with_value(
        grouped_df,
        col,
        (grouped_df["Year"] >= 2015 | ((grouped_df["Year"] == 2014) & (grouped_df["Month"] >= 6)))
    )
    
    # print(id_with_new_value_df)

    merged_df = merge_dfs(
        id_with_old_value_df,
        id_with_new_value_df,
        col
    )
    
    # print(merged_df)

    new_df = replace_value_from_hash_df(new_df, merged_df, col, col)
        
    return new_df
````

````python
# Xử lý trên nhiều cột
def handle(
    df: pd.DataFrame,
    lst_col: list[str]
):
    new_df = df.copy()
    for col in lst_col:
        new_df = handle_each(new_df, col)
    return new_df
````

````python
def concat_dfs(
    df_1: pd.DataFrame,
    df_2: Optional[pd.DataFrame],
):
    res: pd.DataFrame = df_1
    if df_2 is None:
        return res

    res = pd.concat([res, df_2], ignore_index=True).drop_duplicates()
    return res
````

````python
def handle_ignition_factor(df, lst, label='Ignition Factor'):
    stored_old_df: Optional[pd.DataFrame] = None
    stored_new_df: Optional[pd.DataFrame] = None
        
    new_df: pd.DataFrame = df.copy()
    
    for col in lst:
        new_df = handle_each(new_df, col)
        
        grouped_df: pd.DataFrame = new_df.groupby([
            f"{col} ID",
            f"{col} Desc",
            "Year"
        ]).size().reset_index()

        old_tmp_df: pd.DataFrame = get_id_with_value(
            grouped_df,
            col,
            grouped_df["Year"] <= 2013
        )
        old_tmp_df = old_tmp_df.rename(
            columns={
                f"{col} ID": f"{label} ID",
                f"{col} Desc": f"{label} Desc",
            }
        )
        
        stored_old_df = concat_dfs(old_tmp_df, stored_old_df)

        new_tmp_df: pd.DataFrame = get_id_with_value(
            grouped_df,
            col,
            grouped_df["Year"] >= 2013
        )
        
        new_tmp_df = new_tmp_df.rename(
            columns={
                f"{col} ID": f"{label} ID",
                f"{col} Desc": f"{label} Desc",
            }
        )

        stored_new_df = concat_dfs(new_tmp_df, stored_new_df)
        
    merged_df: pd.DataFrame = merge_dfs(
        stored_old_df,
        stored_new_df,
        label
    )
    
    for col in lst:
        new_df = replace_value_from_hash_df(
            new_df,
            merged_df,
            col,
            label,
        )
    
    return new_df
````
